```python
CLIENT_ID = "6fadbdba066e47b8a5eb752de49bb202"
CLIENT_SECRET = "d5fe58aff0094c7586d8542f2fe1f3ac"
client_credentials_manager = SpotifyClientCredentials(client_id=CLIENT_ID, client_secret=CLIENT_SECRET)
sp = spotipy.Spotify(client_credentials_manager = client_credentials_manager)
```

```python
my_playlists = sp.user_playlists('tz0j2laeudcuwymfeh15xovgt')
```

```python
selected_playlists = {}


for playlist in my_playlists['items']:
    if not playlist['collaborative']:
        print(playlist['id'], playlist['name'], playlist['tracks']['total'])
        selected_playlists[playlist['id']] = {
            'name': playlist['name'],
            'number_of_songs': playlist['tracks']['total']
        }

#Eliminating the playlists that I don't need for analysis
eliminate = ['58OroFc8yqiLswqnPHqllM', '2fVTUS27Es6mym3ScOaNUT', '3u2649hgxZu7s6GFeEuSho', 
            '2jOKBcK1uDJiafpTRqKPrX', '5kB3ILLfhrWJl8bIIka7Kg', '10F2GFf0WSnRSyeD4qAL3L',
            '126EqL0ywBsB0ObBXAKY4G', '1Bqstr4Q2xibU6mOnpmxKa']

for playlist in eliminate:
    del selected_playlists[playlist]

selected_playlists
```

```python
columns = ['track_id', 'track_name', 'artists', 'album', 'danceability', 'energy', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'key', 'mode', 'playlist']
All_playlists = pd.DataFrame(columns=columns)
All_playlists
```

```python
for playlist_id in selected_playlists:
    print("Collecting the songs in playlist", selected_playlists[playlist_id]['name'])

    for off in range(0,selected_playlists[playlist_id]['number_of_songs'], 100):
        tracks = sp.playlist_items(playlist_id = playlist_id,limit=100, offset=off)["items"]
        
        for track in tracks:

            artists = ""
            for artist in track['track']['artists']:
                artists += artist['name'] + ", "
                
            song_row = {
                'track_id': track['track']['id'],
                'track_name': track['track']['name'],
                'artists': artists,
                'album': track['track']['album']['name']
            }

            for col in columns[4:-1]:
                song_row[col] = np.nan

            song_row['playlist'] = selected_playlists[playlist_id]['name']

            All_playlists.loc[len(All_playlists.index)] = song_row.values()


            print("==================================================", off, "/", selected_playlists[playlist_id]['number_of_songs'])
    
All_playlists       
```

```python
All_playlists
```

```python
error_track_ids = []

for off in range(0, len(All_playlists.index), 100):
    track_ids = All_playlists['track_id'].iloc[off:100+off].tolist()

    # Remove any empty or invalid track IDs
    track_ids = [tid for tid in track_ids if isinstance(tid, str) and len(tid) == 22]

    try:
        # Filter out None values
        track_ids = [tid for tid in track_ids if tid is not None]

        if track_ids:
            song_features = sp.audio_features(tracks=track_ids)
            
            for feature in song_features:
                track_id = feature['id']
                row_index = All_playlists.index[All_playlists['track_id'] == track_id].tolist()

                if row_index:
                    row = row_index[0]

                    # Update columns 4:-1 with audio features
                    for ele in All_playlists.columns[4:-1]:
                        All_playlists.loc[row, ele] = feature[ele]
    except Exception as e:
        print(f"Error processing track IDs: {e}")
        # Store the failed track IDs
        error_track_ids.extend(track_ids)

# Display the updated DataFrame
print(All_playlists)

# Display the track IDs that encountered errors
print("Track IDs with errors:", error_track_ids)
```
